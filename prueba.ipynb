#### Importar las librerías:

```python
import scratch as red
import numpy as np
import importlib

importlib.reload(red)
```

#### Obtención de datos:
  Obtener los datos de un csv introducido por parámetro, se dividen en 2 arrays, luego se normalizan y se dividen en los datos de entrenamiento y de prueba con un parametro que indica el porcentaje de partición.

```python
porcentaje = 0.8

cel, far = red.getDatos('datos.csv')
celn, farn = red.normalizarDatos(cel, far)
train, test = red.trainTestDataSplit(celn, farn, porcentaje)
```

#### Inicializar las neuronas:
  Se inician las neuronas (numero dado) con pesos aleatorios entre -1 y 1 y un bias de 0.

```python
numero_neuronas = 5

neuronas, capa = red.iniciarNeuronas(numero_neuronas)
```

#### Entrenamiento de las neuronas:
  Se entrenan las neuronas con un $\eta$ (tasa de aprendizaje) y las vueltas que hace el programa.

```python
learning_rate = 0.01
epochs = 10000

red.fit(neuronas, capa, train, learning_rate, epochs)
```

#### Predicción:
  Prediccion de los grados fahrenheit en base a una temperatura en Celsius.

```python
resultado = red.predict(neuronas, capa, 30, cel, far)
print(f"{resultado}ºF")
```
